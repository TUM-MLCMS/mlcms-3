```python
from numpy import *
import matplotlib.pyplot as plt

# prepare plot
fig = plt.figure(figsize=(9,6))
ax1 = fig.add_subplot(1, 1, 1)

def xeq1(r):
    return sqrt(r)

def xeq2(r):
    return -sqrt(r)

domain = linspace(0, 1)
ax1.plot(domain, xeq1(domain), 'b-', label = 'stable state', linewidth = 3)
ax1.plot(domain, xeq2(domain), 'r--', label = 'unstable state', linewidth = 3)
ax1.legend(loc='upper left')

ax1.plot([0], [0], 'go')
ax1.axis([-1, 1, -1.5, 1.5])
ax1.set_xlabel('α')
ax1.set_ylabel('ẋ')
ax1.set_title('Saddle-node bifurcation')
```

```python
from numpy import *
import matplotlib.pyplot as plt

# prepare plot
fig = plt.figure(figsize=(9,6))
ax1 = fig.add_subplot(1, 1, 1)

def xeq1(r):
    return sqrt(r - 3) / sqrt(2.0)

def xeq2(r):
    return -sqrt(r - 3) / sqrt(2.0)

domain = linspace(3, 4)
ax1.plot(domain, xeq1(domain), 'b-', label = 'stable state', linewidth = 3)
ax1.plot(domain, xeq2(domain), 'r--', label = 'unstable state', linewidth = 3)
ax1.legend(loc='upper left')

ax1.plot([0], [0], 'go')
ax1.axis([2, 4, -1.5, 1.5])
ax1.set_xlabel('α')
ax1.set_ylabel('ẋ')
ax1.set_title('Saddle-node bifurcation')
```
